```python
import seaborn as sns
import matplotlib.pyplot as plt
print(supply_chain_data['Delivery Status'].value_counts())
sns.countplot(x='Delivery Status', data=supply_chain_data)
plt.title("Delivery Status Distribution")
plt.show()
```

```python
region_profit = supply_chain_data.groupby('Order Region')['Order Profit Per Order'].sum().sort_values(ascending=False)
print(region_profit)

region_profit.plot(kind='bar', color='skyblue')
plt.title("Profit by Region")
plt.xlabel("Order Region")
plt.ylabel("Total Profit")
plt.show()
```

```python
category_sales = supply_chain_data.groupby('Category Name')['Sales'].sum().sort_values(ascending=False)
print(category_sales)

category_sales.plot(kind='bar', color='coral')
plt.title("Sales by Category")
plt.xlabel("Category Name")
plt.ylabel("Total Sales")
plt.show()
```

```python
shipping_profit = supply_chain_data.groupby('Shipping Mode')['Order Profit Per Order'].sum()
shipping_profit.plot(kind='bar', color='purple')
plt.title("Profit by Shipping Mode")
plt.xlabel("Shipping Mode")
plt.ylabel("Total Profit")
plt.show()
```
